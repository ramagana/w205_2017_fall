### helper functions

```python
def summary(df):
    print("Shape:", df.shape)
    print('Columns:', df.columns)
    return

def unique_report(df):
    for col in df.columns:
        print(col,df[col].nunique(),'values :')
        print(df[col].unique(),'\n\n')
    return

def print_cols(df):
    for col in df.columns:
        print(col,":  ",df[col].nunique(),'distinct values,',df[col].dtype,'data type.')      
    return

# def coerce_to(dfseries,dtype):
#     df.pivot_table(index=[‘id’],
#                              columns=[‘key’],
#                              values=[‘value’],
#                              aggfunc=lambda x: ‘ ‘.join(str(v) for v in x))
#     return

def col_convert(df):
    new_names = [col.lower().replace(" ", "_") for col in df.columns]
    df.columns = new_names
    for i in new_names:
        print(str(i))
    return df
    
def not_avail_fix(df,cols,missing,cast_type):
    for col in cols:
        series =df[col]
        x = np.where(series == missing, np.nan, series).astype(cast_type)
        df[col] = x
    return
```
